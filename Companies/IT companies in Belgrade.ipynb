```python
import undetected_chromedriver as uc
from bs4 import BeautifulSoup 
from re import search
import pandas as pd
import requests
```

```python
def get_parameters(sity):
    if sity == 'Belgrade': parameters = {'locId': 1655, 'locName': 'Belgrade,%20Serbia%20Area', 'num_pages':23}
    if sity == 'Moscow': parameters = {'locId': 1159, 'locName': 'Moscow,%20Russia%20Area', 'num_pages':44}
    if sity == 'Saint Petersburg': parameters = {'locId': 1158, 'locName': 'Saint%20Petersburg,%20Russia%20Area', 'num_pages':20 }
    return parameters
```

```python
def get_companies(parameters):
    companies = []
    driver = uc.Chrome()
    for page in range(1, parameters['num_pages']+1):
        driver.get('https://www.glassdoor.com/Explore/browse-companies.htm?overall_rating_low=3&sector=10013&locType=M&page={i}\
        &locId={my_dict[locId]}&locName={my_dict[locName]}'.format(i = page, my_dict=parameters))
        HTML = driver.page_source
        soup = BeautifulSoup(HTML , 'html.parser')
        company_information = soup.find_all('span', class_ ="align-items-center mb-xsm")
        for company_name_and_rating in company_information:
            company_name = search(r'(\D+)', company_name_and_rating.text).group(1) 
            companies.append(company_name)
    return companies
```

```python
parameters_for_Belgrade = get_parameters('Belgrade')
parameters_for_Moscow = get_parameters('Moscow')
parameters_for_Saint_Petersburg = get_parameters('Saint Petersburg')
```

```python
companies_in_Belgrade =  pd.DataFrame(get_companies(parameters_for_Belgrade))
companies_in_Saint_Petersburg =  pd.DataFrame(get_companies(parameters_for_Moscow))
companies_in_Moscow =  pd.DataFrame(get_companies(parameters_for_Saint_Petersburg))
```

```python
companies_in_Moscow_or_Saint_Petersburg = pd.merge(companies_in_Moscow, companies_in_Saint_Petersburg,  how='outer').drop_duplicates()
companies_in_MSK_or_Petersburg_and_Belgrade  = pd.merge(companies_in_Moscow_or_Saint_Petersburg, companies_in_Belgrade, how='inner')
```
